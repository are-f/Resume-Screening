# Now Cleaning the df[Category] part

```python
import re
def cleanResume(resumeText):
    resumeText = re.sub('http\S+\s*', ' ', resumeText)  # remove URLs
    resumeText = re.sub('RT|cc', ' ', resumeText)  # remove RT and cc
    resumeText = re.sub('#\S+', '', resumeText)  # remove hashtags
    resumeText = re.sub('@\S+', '  ', resumeText)  # remove mentions
    resumeText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ',resumeText) #remove punctuations
    resumeText = re.sub(r'[^\x00-\x7f]',r' ', resumeText) 
    resumeText = re.sub('\s+', ' ', resumeText)  # remove extra whitespace
    return resumeText
    
df['cleaned_resume'] = df.Resume.apply(lambda x: cleanResume(x))
```

# Now Import Nltk lib

```python
import nltk
from nltk.corpus import stopwords
import string
from wordcloud import WordCloud
```

```python
# Download the necessary NLTK data
nltk.download('stopwords')
nltk.download('punkt')  # Ensure 'punkt' is downloaded for tokenization

oneSetOfStopWords = set(stopwords.words('english') + ['``', "''"])
totalWords = []
Sentences = df['Resume'].values  
cleanedSentences = ""

for records in Sentences:
    cleanedText = cleanResume(records)  # Ensure cleanResume function is defined
    cleanedSentences += cleanedText
    from nltk.tokenize import TreebankWordTokenizer
    tokenizer = TreebankWordTokenizer()
    requiredWords = tokenizer.tokenize(cleanedText)
  # Tokenize the cleaned text
    for word in requiredWords:
        if word not in oneSetOfStopWords and word not in string.punctuation:
            totalWords.append(word)
    
# Get the frequency distribution of the words
wordfreqdist = nltk.FreqDist(totalWords)
mostcommon = wordfreqdist.most_common(50)
print(mostcommon)
```

# Now Train test split and fit transform

```python
from sklearn.model_selection import train_test_split      #importing required libs
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack

requiredText = df['cleaned_resume'].values
requiredTarget = df['Category'].values

word_vectorizer = TfidfVectorizer(  # This object converts the raw text into TF-IDF vectors, which are numerical representations of the words based on their importance in the corpus.
    sublinear_tf=True,              #This parameter enables sublinear term frequency scaling, which applies logarithmic scaling to the term frequency. This makes the high-frequency words less dominant.
    stop_words='english')           #Automatically removes common English stopwords (e.g., "the", "and", etc.) during vectorization to focus on more meaningful words.
word_vectorizer.fit(requiredText)   #Fitting required data
WordFeatures = word_vectorizer.transform(requiredText)    #Data transformation

print ("Feature completed .....")
```

```python
# train_test split in the ratio 80:20
X_train,X_test,y_train,y_test = train_test_split(WordFeatures,requiredTarget,random_state=1, test_size=0.2,shuffle=True, stratify=requiredTarget)
print(X_train.shape)
print(X_test.shape)
```

# Now Import Important Libs

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.naive_bayes import MultinomialNB
from sklearn.multiclass import OneVsRestClassifier
from sklearn import metrics
from sklearn.metrics import accuracy_score
from pandas.plotting import scatter_matrix
from sklearn.neighbors import KNeighborsClassifier
```

# Now Fit and Prediict the model

```python
clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train, y_train)
prediction = clf.predict(X_test)
print('Accuracy of KNeighbors Classifier on training set: {:.2f}'.format(clf.score(X_train, y_train)))
print('Accuracy of KNeighbors Classifier on test set:     {:.2f}'.format(clf.score(X_test, y_test)))
```

# Now Check Model Score

```python
print("\n Classification report for classifier %s:\n%s\n" % (clf, metrics.classification_report(y_test, prediction)))
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Function to match resumes against a job description
def match_resumes_to_job(job_description, resumes_df, vectorizer):
    # Clean the job description
    cleaned_job_desc = cleanResume(job_description)
    
    # Transforming the job description using the TF-IDF vectorizer
    job_desc_vector = vectorizer.transform([cleaned_job_desc])
    
    # Calculating similarity scores with all resumes
    similarity_scores = cosine_similarity(job_desc_vector, resumes_df['vector'].tolist()) #computes the cosine similarity between vectors
    
    # Creating DataFrame to store the results
    similarity_df = pd.DataFrame(similarity_scores.T, columns=['Similarity'])
    similarity_df['Resume'] = resumes_df['Resume'].values
    similarity_df.sort_values(by='Similarity', ascending=False, inplace=True)
    
    return similarity_df.head(10)  # Return top 10 matches

# Main program
if __name__ == "__main__":
    # Loading dataset with resumes and their vectors
    df = pd.read_csv('C:\\Users\\pc\\Downloads\\Data science\\UpdatedResumeDataSet.csv')
    
    df['vector'] = [word_vectorizer.transform([cleanResume(resume)]).toarray().flatten() for resume in df['Resume']]
    
    # Input job description from the user
    job_description = input("Job description: ")
    
    # Match resumes to the job description
    top_matches = match_resumes_to_job(job_description, df, word_vectorizer)
    
    # Display the top matching resumes
    print("\nTop 10 Matching Resumes:")
    print(top_matches)
```
